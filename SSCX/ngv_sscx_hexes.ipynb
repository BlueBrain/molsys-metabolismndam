```python
#ngv_sscx_id_map_d = pd.Series(ngv_sscx_id_map.ngv.values,index=ngv_sscx_id_map.sscx).to_dict()
ngv_sscx_id_map_d = pd.Series(df.loc[:,'node_ids'].values,index=df.loc[:,'index'].values).to_dict()
```

```python
# Use Sirio's hexes target

with open("/gpfs/bbp.cscs.ch/project/proj34/metabolismndam/SSCX/hexes.target","r") as hf:
    
    hexes = {}
    
    startl = False
    endl = False
    
    for line in hf:
        
        if "Target Cell" in line:
            print(line)
            hexes_k = line.replace("Target Cell ","").rstrip("\n").replace(" ","")
            
            startl = True
            endl = False
            continue
        elif "}" in line:
            startl = False
            endl = True
            continue
            
        elif (startl == True) and (endl == False):
            if line != "{\n":
                if not hexes_k in hexes:
                    print(line.count("a"))
                    hexes[hexes_k] = [i.replace("a","").replace("\n","") for i in line.rstrip("\n").split(" ") if i != '']
                    print(hexes_k,len(hexes[hexes_k]))
                else:
                    print("ATTENTION: ",hexes_k)
```

```python
#sscxids = set(list(ngv_sscx_id_map.loc[:,'sscx'].values))
sscxids = set(list(df.loc[:,'index'].values))
```

```python
hex_ngv = []
hexes_allids = []

for k1,v in hexes.items():
    
    if not k1 in ["hex_O1","hex_O2","hex_all"]:
        for k in v:
            hexes_allids.append(int(k))
            
            if int(k) in ngv_sscx_id_map_d:
                hex_ngv.append(int(k))
    
print(len(set(hex_ngv)), len(set(hexes_allids)), len(hexes_allids))
```

```python
ngv_hex_counts = {}
ngv_hex_ids = {}

for k1,v in hexes.items():
    if not k1 in ["hex_O1","hex_O2","hex_all"]:
        
        if len(set(hex_ngv).intersection(set([int(i) for i in v]))) != 0:
            
            print(f"{k1}, ngv: {len(set(hex_ngv).intersection(set([int(i) for i in v])))}, sscx: {len(set(v))}") 

            ngv_hex_counts[k1] = len(set(hex_ngv).intersection(set([int(i) for i in v])))
            
            ngv_hex_ids[k1] = [ ngv_sscx_id_map_d[i] for i in set(hex_ngv).intersection(set([int(i) for i in v]))]
```
